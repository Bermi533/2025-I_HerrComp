```python
import numpy as np 
import matplotlib.pyplot as plt 
```

## Movimiento Browniano

```python
x=0
y=0
X_pos=[x]
Y_pos=[y]
S=1000
Deltal=0.1
for t in range(S):
    theta = 2*np.pi*np.random.rand()
    x+=Deltal*np.cos(theta)
    y+=Deltal*np.sin(theta)
    X_pos.append(x)
    Y_pos.append(y)
plt.plot(X_pos, Y_pos, 'r-',linewidth=0.3)
plt.show()
```

## Vuelos de Lvy

Ahora L no es fijo, usando distribucin de Pareto

```python
for j in range(10):
    x=0
    y=0
    X_pos=[x]
    Y_pos=[y]
    S=10000
    for t in range(S):
        Deltal=np.random.pareto(1.5)
        theta = 2*np.pi*np.random.rand()
        x+=Deltal*np.cos(theta)
        y+=Deltal*np.sin(theta)
        X_pos.append(x)
        Y_pos.append(y)
    plt.plot(X_pos, Y_pos, linewidth=0.5)
plt.show()
```
