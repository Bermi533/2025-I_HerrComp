```python
import numpy as np 
import scipy as sci 
import matplotlib.pyplot as plt 
```

```python
def phi2(x):
    return np.sin(x)*np.exp(x**2)

def phi(a,x):
    return a*x

def G(e,x):
    if e <= x:
        out = (x-1)*e
    else:
        out = (e-1)*x
    return out

def I(e,x):
    return G(e,x)*phi(2,e)

def u(x):
    if x == 0 or x == 1:
        out = 0
    else:
        def I_(e):
            return I(e,x)
        out = sci.integrate.quad(I_, 0, 1)[0]
    return out 

X = np.linspace(0, 1, 100)
U = np.array([u(x) for x in X])
      
plt.plot(X, U, label='u(x)')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Plot of u(x), with Green function')
plt.legend()
plt.grid()
plt.show()
```
