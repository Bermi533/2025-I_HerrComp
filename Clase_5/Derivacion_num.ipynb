$\textbf{Derivación numérica}  \\$

f(x)-> f'(x) $ \\$
f'(x) $ \approx \frac{f(x+h) - f(x)}{h} \to h << \\$
¿El error con respecto a h de la cosa de arriba de cuánto es? $\\$
$f(x+h) = f(x)+f'(x)h+\frac{1}{2}f"(x)h^{2} + ... \\$
Despejando f'(x): $\\$
$f´(x) = \frac{f(x+h) - f(x)}{h} - \frac{1}{2} f"(x) h +...$ Es decir, el error ed de orden h (O(h)) $\\$
O sea, entre más pequeño sea h, mejor $\\$

Hay otra definición para la derivada, vamos a deducirla: $\\$
$f(x+h)=f(x)+f'(x)h+\frac{1}{2}f"(x)h^{2} + O(h^{3})\\
f(x-h)=f(x)-f'(x)h+\frac{1}{2}f"(x)h^{2} + O(h^{3})\\$
Por lo tanto: $\\$
$f(x+h)-f(x-h) = 2f'(x)h + O(h^{3}) \\$
$f'(x) = \frac{f(x+h)-f(x-h)}{2h} + O(h^{2})$ Luego, si h es pequeño $h^{2}<h$

```python
#Derivar f(x) = sen(x)
import numpy as np  
import matplotlib.pyplot as plt 

h = 0.01

def f(x):
    return np.sin(x)

def f1(x):
    return np.cos(x)

def Der_1(x, h):
    return (f(x+h)-f(x))/h

def Der_2(x, h):
    return (f(x+h)-f(x-h))/(2*h)

X = []
D1 = []
D2 = []
cos = []

x = 0
while x <= 4*np.pi:
    X.append(x)
    D1.append(Der_1(x,h))
    D2.append(Der_2(x,h))
    cos.append(f1(x))
    x += 0.001
```

```python
plt.plot(X,D1,'-b')
plt.plot(X,D2,':r')
plt.plot(X,cos, 'g')
```
