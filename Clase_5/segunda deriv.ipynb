$f"(x) = \frac{f(x+h) + f(x-h) -2 f(x)}{h^{2}} + O(h²) $ La demostración está en la libreta

```python
#Ejercicio 12 página 196 -> numerical methods in engineering with python 3
import numpy as np
import matplotlib.pyplot as plt

R = 0.09 #m
theta_punto = 5000*(2*np.pi)/60 #Hz
alpha = 2.5
h = 0.001

#Hallar x"(t)
def x(theta):
    return R*(np.cos(theta) + np.sqrt(alpha**2 - (np.sin(theta))**2))

def x_dosp(theta, h):
    return ((x(theta+h) + x(theta-h) - 2*x(theta))/(h**2))

def a_x(t):
    theta = theta_punto*t
    return x_dosp(theta, h)*theta_punto**2

#theta = theta_incial + theta_punto* t
#Haciendo una lista con posibles valores para theta en función del tiempo:

T = np.linspace(0,0.05,10001)
A = []

for t in T:
    A.append(a_x(t))
```

```python
plt.plot(T,A,'g-')
plt.xlabel('Tiempo (s)', size = 15)
plt.ylabel(r'$a_{x}(t)$ $(m/s^2)$', size = 15)
plt.grid()
plt.show()
```
