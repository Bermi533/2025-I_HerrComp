```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def coin():
    x = np.random.rand()
    if x <= 0.5:
        out  =  -1
    else:
        out =  1 
    return out
```

```python
Position = []
pos = 0
for t in range(10000):
    pos += coin()
    Position.append(pos)
```

```python
#Position
```

```python
plt.plot(Position, 'g')
plt.xlabel('Tiempo t', size = 10)
plt.ylabel('Posición x(t)', size = 10)
plt.show()
```

```python
for i in range(100):
    Position = []
    pos = 0
    for t in range(10000):
        pos += coin()
        Position.append(pos)
    plt.plot(Position)   
plt.xlabel('Tiempo [t]', size = 10)
plt.ylabel('Posición [x(t)]', size = 10)
plt.show()
```
