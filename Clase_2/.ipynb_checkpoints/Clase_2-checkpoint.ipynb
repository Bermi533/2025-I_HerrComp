```python
import numpy as np
```

```python
def Pi(N):
  n = 0
  for i in range (N):
    x = np.random.rand()
    y = np.random.rand()
    r = x**2 + y**2
    if r <= 1:
      n += 1
  return 4*(n/N)
```

```python
N = 10000
Data = []
for a in range(1000):
  val = Pi(N)
  Data.append(val)
```

```python
#Data
```

```python
np.mean(Data), np.std(Data)
```

```python
pi_mio = sum(Data)/len(Data)
pi_py = np.pi
print(pi_mio)
print(pi_py)
```

```python
dpr = (np.abs(pi_mio-pi_py)/pi_py)*100
print(dpr)
```
