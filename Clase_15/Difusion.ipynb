```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
mpl.rcParams['text.usetex'] = False 
```

```python
Delta_x=0.01
X=np.arange(-1,1+Delta_x,Delta_x)

C=1
Delta_t=Delta_x**2/C

T=np.arange(0,0.1+Delta_t,Delta_t)
```

```python
A,B=np.meshgrid(T,X)
```

```python
Psi=np.zeros(A.shape)
```

```python
plt.figure(figsize=(6,2.5))
plt.imshow(Psi,extent=(min(T),max(T),min(X),max(X)),aspect=0.02)
plt.ylabel(r'$x$',size=20)
plt.xlabel(r'$t$',size=20)
plt.colorbar().set_label(label=r'$\Psi(x,t)$',size=20)
plt.show()
```

```python
m=int((len(X)-1)/2)
Psi[m][0]=1/(2*Delta_x)
Psi[m+1][0]=1/(4*Delta_x)
Psi[m-1][0]=1/(4*Delta_x)
```

```python
plt.plot(Psi[::,0],'k')
plt.show()
```

```python
plt.figure(figsize=(6,2.5))
plt.imshow(Psi,extent=(min(T),max(T),min(X),max(X)),aspect=0.02)
plt.ylabel(r'$x$',size=20)
plt.xlabel(r'$t$',size=20)
plt.colorbar().set_label(label=r'$\Psi(x,t)$',size=20)
plt.show()
```

```python
for i in range(1,len(X)-1):
    Psi[i][1]=0.5*(Psi[i+1][0]+Psi[i-1][0])
```

```python
plt.plot(Psi[::,0],'k')
plt.plot(Psi[::,1],'r')
plt.show()
```

```python
for j in range(1,len(T)):
    for i in range(1,len(X)-1):
        Psi[i][j]=0.5*(Psi[i+1][j-1]+Psi[i-1][j-1])
```

```python
plt.figure(figsize=(6,2.5))
plt.imshow(Psi,extent=(min(T),max(T),min(X),max(X)),aspect=0.02,vmax=3)
plt.plot(T,3*np.sqrt(C*T),'w--')
plt.plot(T,-3*np.sqrt(C*T),'w--')
plt.ylabel(r'$x$',size=20)
plt.xlabel(r'$t$',size=20)
plt.colorbar(extend='max').set_label(label=r'$\Psi(x,t)$',size=20) #extend=max, para decir que el color amarillo es tres o más
plt.show()
```

```python
def coin():
    x = np.random.rand()
    if x <= 0.5:
        out  =  -Delta_x
    else:
        out =  Delta_x
    return out

Position = []
pos = 0
for t in range(100):
    pos += coin()
    Position.append(pos)
```

```python
for i in range(100):
    Position = []
    pos = 0
    for t in range(1000):
        pos += coin()
        Position.append(pos)
    plt.plot(Position)   
plt.xlabel('Tiempo [t]', size = 10)
plt.ylabel('Posición [x(t)]', size = 10)
plt.show()
```

```python
for rr in range(1000):
    x=0
    X_pos=[x]
    for t in T[1::]:
        x+=np.random.choice([-Delta_x,Delta_x], p=[0.5,0.5])
        X_pos.append(x)
    plt.plot(T,X_pos, linewidth=0.1)
plt.plot(T,3*np.sqrt(C*T),'k--')
plt.plot(T,-3*np.sqrt(C*T),'k--')
plt.ylabel(r'$x$', size=20)
plt.xlabel(r'$t$', size=20)
plt.show()
```

```python
t_m=0.04
Data=[]
for rr in range(10000):
    x=0
    t=0
    while t<=t_m:
        x+=np.random.choice([-Delta_x,Delta_x],p=[0.5,0.5])
        t+=Delta_t
    Data.append(x)
```

```python
plt.plot(Data,'k.')
plt.show()
```

```python
value,base=np.histogram(Data,density=True,bins=100)
plt.plot(X,Psi[::,400],'m-')
plt.plot(base[1::],value,'c.')
plt.show()
```
