```python
import numpy as np   
import matplotlib.pyplot as plt   
```

```python
#Examinar la solución del modelo de Kuramoto para 0<=t<=15
matL = [[2,-1,-1],[-1,2,-1],[-1,-1,2]]
theta_0 = [np.pi, np.pi/2, 5*np.pi/4]
matL_ = np.array(np.array(matL)*(-1))

eigvals, Q = np.linalg.eig(matL_)
Qinv = np.linalg.inv(Q)
matLambda= np.diag(np.exp(eigvals))

Data = []
Time = []

def Temp_ev(t):
    Lambda_exp = np.diag(np.exp(eigvals*t))
    return np.matmul(np.matmul(Q,Lambda_exp), Qinv)

t = 0
while t <= 15:
    Data.append(np.matmul(Temp_ev(t),theta_0))
    Time.append(t)
    t += 0.01
    
Data = np.array(Data)

plt.semilogx(Time,Data.T[0])
plt.semilogx(Time,Data.T[1])
plt.semilogx(Time,Data.T[2])
plt.show()

    
```
