```python
import numpy as np  
import matplotlib.pyplot as plt 
```

y" = -(19/4)y-10y' $ \\  $
con y(0) = -9, y'(0) = 0

```python
matL_ = np.array([[0, 1], [-19/4, -10]])
y_0 = np.array([[-9], [0]])

eigvals, Q = np.linalg.eig(matL_)
Qinv = np.linalg.inv(Q)
matLambda= np.diag(np.exp(eigvals))

Data = []
Datateo = []
X = []

def y(x):
    Lambda_exp = np.diag(np.exp(eigvals*x))
    return np.matmul(np.matmul(Q,Lambda_exp), Qinv)

def yteo(x):
    a = -(19/2)*np.exp(-x/2)
    b = (1/2)*np.exp(-19*x/2)
    return a + b

x = 0
while x <= 15:
    Data.append(np.matmul(y(x),y_0))
    Datateo.append(yteo(x))
    X.append(x)
    x += 0.01
    
Y = np.array(Data)
Yteo = np.array(Datateo)


plt.plot(X,Y[:,0,0], "-b", label = "Solución con métodos vistos en clase")
plt.plot(X,Yteo, "--r", label = "Solución esperada") #alternativamente plt.semilogx 
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.show()

```
