```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Zorros y conejos

#Par√°metros de las interacciones entre especies
alpha=1.1
beta=0.4
gamma=0.4
delta=0.1

#Ecuaciones de Lotka-Volterra
def f(t,x,y):
    return alpha*x-beta*x*y

def g(t,x,y):
    return delta*x*y-gamma*y
```

```python
#RK cuarto orden para ecs. acopladas
T=[]
X=[]
Y=[]

t=0
x=20
y=6

h=0.01
while t<=100:
    T.append(t)
    X.append(x)
    Y.append(y)
    k1x=f(t,x,y)
    k1y=g(t,x,y)
    k2x=f(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
    k2y=g(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
    k3x=f(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
    k3y=g(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
    k4x=f(t+h,x+k3x*h,y+k3y*h)
    k4y=g(t+h,x+k3x*h,y+k3y*h)
    x+=(1.0/6)*(k1x+2*k2x+2*k3x+k4x)*h
    y+=(1.0/6)*(k1y+2*k2y+2*k3y+k4y)*h
    t+=h
```

```python
plt.plot(T,X,'k--', label='Conejos')
plt.plot(T,Y,'r-', label='Zorros')
plt.xlim(0,100)
plt.show()
```
