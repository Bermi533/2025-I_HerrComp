```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def f(x,y):
    return (1+2*x)*y**(np.sin(x*y))
```

```python
#RK cuarto orden
H=[0.5,0.1,0.01,0.001]

def RK4(x0,y0,h,xmax):
    X=[]
    Y=[]
    x=x0
    y=y0
    while x<=xmax:
        X.append(x)
        Y.append(y)
        k1 = f(x,y)
        k2 = f(x+(1/2)*h, y+(1/2)*k1*h)
        k3_ = f(x+(1/2)*h,y+(1/2)*h*k2)
        k4 = f(x+h,y+k3_*h)
        y += (1/6)*(k1+2*k2+2*k3_+k4)*h
        x += h
    return X,Y
     
```

```python
X1,Y1=RK4(0,1,0.5,3)
X2,Y2=RK4(0,1,0.1,3)
X3,Y3=RK4(0,1,0.01,3)
X4,Y4=RK4(0,1,0.001,3)
```

```python
plt.semilogy(X1,Y1,':k',label='h=0.5')
plt.semilogy(X2,Y2,':g',label='h=0.1')
plt.semilogy(X3,Y3,':b',label='h=0.01')
plt.semilogy(X4,Y4,':r',label='h=0.001')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('RK4 para diferentes h')
plt.grid()
plt.legend()
plt.show()
```
