```python
#Runge-Kutta de segundo orden, Ej: método del punto medio con y' = (1+2x)sqrt(y) entre 0 y 1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
    return (1+2*x)*np.sqrt(y)
```

```python
X = []
Y_euler=[]

x=0
y=1
h=0.01

while x<=1:
    X.append(x)
    Y_euler.append(y)
    y+=f(x,y)*h #Método de Euler (primer orden) y_(i+q)=y_i + f(x_i,y_i)*h
    x+=h        #x_(i+1)=x_i + h
```

```python
plt.plot(X,Y_euler,'k.')
plt.show()
```

```python
#ahora sí hagamos rk2 con método de punto medio
X = []
Y_RK2=[]

x = 0
y = 1
h = 0.01

while x<=1:
    X.append(x)
    Y_RK2.append(y)
    k1 = f(x,y)
    k2 = f(x+(1/2)*h, y+(1/2)*k1*h)
    y+=k2*h
    x+=h

```

```python
plt.plot(X,Y_RK2,'r-')
plt.plot(X,Y_euler,'k--')
plt.show()
```

```python
#tercer orden
X = []
Y_RK3=[]

x = 0
y = 1
h = 0.01

while x<=1:
    X.append(x)
    Y_RK3.append(y)
    k1 = f(x,y)
    k2 = f(x+(1/2)*h, y+(1/2)*k1*h)
    k3 = f(x+h,y-k1*h+2*k2*h)
    y += (1/6)*(k1+4*k2+k3)*h
    x += h
```

```python
plt.plot(X,Y_euler,'k--', label='Primer orden')
plt.plot(X,Y_RK2,'r-', label='Segundo orden')
plt.plot(X,Y_RK3,'g:', label='Tercer orden')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Comparación de métodos Runge-Kutta')
plt.grid()
plt.legend()
plt.show()
```

```python
#RK cuarto orden
X=[]
Y_RK4=[]

x=0
y=1
h=0.01

while x<=1:
    X.append(x)
    Y_RK4.append(y)
    k1 = f(x,y)
    k2 = f(x+(1/2)*h, y+(1/2)*k1*h)
    k3_ = f(x+(1/2)*h,y+(1/2)*h*k2)
    k4 = f(x+h,y+k3_*h)
    y += (1/6)*(k1+2*k2+2*k3_+k4)*h
    x += h
    
```

```python
plt.plot(X,Y_euler,'k--', label='Primer orden')
plt.plot(X,Y_RK2,'r-', label='Segundo orden')
plt.plot(X,Y_RK3,'g:', label='Tercer orden')
plt.plot(X,Y_RK4,'b:', label='Cuarto orden')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Comparación de métodos Runge-Kutta')
plt.grid()
plt.legend()
plt.show()
```

```python
def Sol_y(x):
    return (0.5*x+0.5*x**2+1)**2
```
