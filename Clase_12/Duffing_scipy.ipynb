```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from scipy.integrate import solve_ivp #initial value problem
```

```python
#Vamos a hacerlo ahora con el oscilador de Duffing
delta=0.1*np.pi
alpha=0
beta=4*np.pi**2
gamma=30*np.pi**2
omega=2*np.pi
```

```python
def f(t,y):
    Fvec=np.zeros(2)
    Fvec[0]=y[1]
    Fvec[1]=gamma*np.cos(omega*t)-alpha*y[0]-delta*y[1]-beta*y[0]**3
    return Fvec
```

```python
t_span=[0,100]
t_eval=np.linspace(0,100,1001)
y0=[1,0]
sol=solve_ivp(f, t_span, y0,t_eval=t_eval)
```

```python
T=sol.t
Y1=sol.y[0]
```

```python
plt.plot(T, Y1, '-b')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.show()
```

```python
t_span=[0,10000]
t_eval=np.linspace(0,10000,10001)
y0=[0,0]
sol=solve_ivp(f, t_span, y0,t_eval=t_eval)
```

```python
T=sol.t
X=sol.y[0]
V=sol.y[1]
```

```python
#plt.plot(X,V,'g--')
plt.plot(X,V,'k.', markersize=0.3)
plt.title('Espacio de fase', size=15)
plt.xlabel(r'$x$', size=20)
plt.ylabel(r'$\dot{x}$', size=20)
plt.grid()
plt.show()
#Sección de Poincaré
```
