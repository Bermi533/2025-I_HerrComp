```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from scipy.integrate import solve_ivp #initial value problem
```

```python
def f(t,y):
    return y
```

```python
t_span=[0,5]
t_eval=np.linspace(0,5,1000)
y0=[1]
sol=solve_ivp(f, t_span, y0,t_eval=t_eval)
```

```python
T=sol.t
Y=sol.y[0]
```

```python
plt.plot(T, Y, '-r', label='Sol. numérica')
plt.plot(T, np.exp(T), '--k', label='Sol. analítica')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid()
plt.show()
```

```python
alpha=1
beta=0.1
gamma=0.5
delta=0.1

def f0(t,y):
    Fvec=np.zeros(2)
    Fvec[0]=alpha*y[0]-beta*y[0]*y[1]
    Fvec[1]=-gamma*y[1]+delta*y[0]*y[1]
    return Fvec
```

```python
t_span0=[0,100]
t_eval0=np.linspace(0,100,1001)
y00=[50,10]
sol0=solve_ivp(f0, t_span0, y00,t_eval=t_eval0)
```

```python
T0=sol0.t
Y1=sol0.y[0]
Y2=sol0.y[1]
```

```python
plt.plot(T0, Y1, '-r', label='Zorros')
plt.plot(T0, Y2, '--k', label='Conejos')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid()
plt.show()
```
