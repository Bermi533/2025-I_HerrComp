```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from scipy.integrate import solve_ivp #initial value problem
```

```python
import networkx as nx
```

```python
Nv=10
G=nx.star_graph(Nv) #grafo estrella
matA=nx.adjacency_matrix(G).todense()  #matriz de cómo está conectado todo, matriz de adyacencia
```

```python
matA
```

```python
def f(t,y): #y son las fases de los diferentes osciladores
    Fvec=np.zeros(Nv)
    for i in range(Nv):
        Fvec[i]=sum(matA[i][j]*np.sin(y[j]-y[i]) for j in range(Nv))
    return Fvec
```

```python
t_span=[0,10]
t_eval=10**np.linspace(-3,1,100)
y0=np.zeros(Nv)
y0[0]=np.pi/2

sol=solve_ivp(f,t_span,y0,t_eval=t_eval)
```

```python
T=sol.t
for i in range(Nv):
    plt.semilogx(T, sol.y[i])
plt.xlabel(r'$t$', size=20)
plt.ylabel(r'$\theta_i(t)$', size=20)
plt.grid()
plt.show()
```

El oscilador principal empieza en pi/2 y decae con el tiempo. EL resto de osciladores se interpretan en la curva cyan.

```python
t_span=[0,10]
t_eval=10**np.linspace(-3,1,100)
y0=np.random.rand(Nv) #fases aleatorias
sol=solve_ivp(f,t_span,y0,t_eval=t_eval)
```
