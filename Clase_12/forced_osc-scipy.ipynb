```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from scipy.integrate import solve_ivp #initial value problem
```

```python
#Vamos a hacerlo ahora con el oscilador forzado
m=1
gamma=0.1
k=1
omega=2*np.pi
F0=1
```

```python
def f(t,y):
    Fvec=np.zeros(2)
    Fvec[0]=y[1]
    Fvec[1]=1.0/m * (F0*np.cos(omega*t)-k*y[0]-gamma*y[1])
    return Fvec
```

```python
t_span=[0,100]
t_eval=np.linspace(0,100,1001)
y0=[0,0]
sol=solve_ivp(f, t_span, y0,t_eval=t_eval)
```

```python
T=sol.t
Y1=sol.y[0]
```

```python
plt.plot(T, Y1, '-r')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.show()
```
