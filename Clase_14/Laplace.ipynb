```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
delta_x=0.01
delta_y=0.01

X=np.arange(0,1+delta_x,delta_x)
Y=np.arange(0,1+delta_y,delta_y)
```

```python
A,B=np.meshgrid(X,Y)
Psi_sol=np.zeros(A.shape)
```

```python
A.shape
```

```python
plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$y$',size=20)
plt.show()
```

```python
Psi_sol[::,0]=100 #Para todas las filas, la primera columna. Pone una de las primeras condiciones iniciales 
plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$y$',size=20)
plt.show()
```

```python
Psi_sol[-1,::]=100 #Para la última fila, todas las columnas
plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$y$',size=20)
plt.show()
```

```python
Psi_sol[::,-1]=100 #Para la última columna todas las filas
plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$y$',size=20)
plt.show()
```

```python
Psi_new=np.copy(Psi_sol)
for i in range(1,len(X)-1):
    for j in range(1,len(Y)-1):
        Psi_new[i][j]=(1.0/4)*(Psi_sol[i+1][j]+Psi_sol[i][j+1]+Psi_sol[i-1][j]+Psi_sol[i][j-1])
```

```python
plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$y$',size=20)
plt.xticks([],[])
plt.yticks([],[])
plt.show()
```

```python
for n in range(1000):
    Psi_new=np.copy(Psi_sol)
    for i in range(1,len(X)-1):
        for j in range(1,len(Y)-1):
            Psi_new[i][j]=(1.0/4)*(Psi_sol[i+1][j]+Psi_sol[i][j+1]+Psi_sol[i-1][j]+Psi_sol[i][j-1])
    Psi_sol=Psi_new        
```

```python
#Difusión de las fronteras hacia dentro
plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$y$',size=20)
plt.xticks([],[])
plt.yticks([],[])
plt.show()
```

```python
#Verificar que sí es la solución de Laplace
Psi_sol=np.zeros(A.shape)
Psi_sol[::,0]=100
Psi_sol[-1,::]=100
Psi_sol[::,-1]=100

Error=[]

for n in range(10000):
    Psi_new=np.copy(Psi_sol)
    for i in range(1,len(X)-1):
        for j in range(1,len(Y)-1):
            Psi_new[i][j]=(1.0/4)*(Psi_sol[i+1][j]+Psi_sol[i][j+1]+Psi_sol[i-1][j]+Psi_sol[i][j-1])
    Error.append(np.mean(np.abs(Psi_new-Psi_sol).flatten())) #.flatten() para convertir la matriz en una sola lista       
    Psi_sol=Psi_new
```

```python
plt.semilogy(Error)
plt.show()
```

```python
plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$y$',size=20)
plt.xticks([],[])
plt.yticks([],[])
plt.show()
```

```python
#Jugando con las condiciones
Psi_sol=np.zeros(A.shape)
z=np.linspace(0,2*np.pi,101)
Psi_sol[::,0]=np.sin(z)
Psi_sol[-1,::]=100
Psi_sol[::,-1]=15

Error=[]

for n in range(10000):
    Psi_new=np.copy(Psi_sol)
    for i in range(1,len(X)-1):
        for j in range(1,len(Y)-1):
            Psi_new[i][j]=(1.0/4)*(Psi_sol[i+1][j]+Psi_sol[i][j+1]+Psi_sol[i-1][j]+Psi_sol[i][j-1])
    Error.append(np.mean(np.abs(Psi_new-Psi_sol).flatten())) #.flatten() para convertir la matriz en una sola lista       
    Psi_sol=Psi_new

plt.imshow(Psi_sol)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$y$',size=20)
plt.xticks([],[])
plt.yticks([],[])
plt.show()
```
