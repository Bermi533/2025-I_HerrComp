```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp
```

```python
k=2.0*np.pi
def fun(x, y):
    return np.array([y[1], -k**2*y[0]])
```

```python
def bc(ya, yb):
    return np.array([ya[0], yb[0]])
```

```python
x = np.linspace(0, 1, 100)

y_initial_g = np.zeros((2,x.size))
y_initial_g[0][1] = 100
```

```python
res_a = solve_bvp(fun,bc,x,y_initial_g)
```

```python
res_a
```

```python
plt.figure(figsize=(5,4))
plt.plot(res_a.x,res_a.y[0])
plt.xlabel(r'$x$',size=15)
plt.ylabel(r'$X(x)$',size=15)
plt.show()
```
