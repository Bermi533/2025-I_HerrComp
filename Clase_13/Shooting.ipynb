```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import solve_ivp
```

```python
def fun(t,y):
    return np.array([y[1],-3*y[0]])
```

```python
t_span=[0,2*np.pi]
u=0.1
y0=[7,u]
t_eval=np.linspace(0,2*np.pi,101)
sol=solve_ivp(fun,t_span,y0,t_eval=t_eval)
```

```python
T=sol.t
X=sol.y[0]
plt.plot(T,X,'k-')
plt.plot(t_span[1],X[-1],'ro')
plt.xlabel(r'$t$',size=20)
plt.ylabel(r'$x(t)$',size=20)
plt.show()
```

```python
U_list=np.linspace(-3,3,101)
for u in U_list:
    t_span=[0,2*np.pi]
    y0=[7,u]
    t_eval=np.linspace(0,2*np.pi,101)
    sol=solve_ivp(fun,t_span,y0,t_eval=t_eval)
    T=sol.t
    X=sol.y[0]
    plt.plot(T,X,'-')
    plt.plot(t_span[1],X[-1],'o')
plt.xlabel(r'$t$',size=20)
plt.ylabel(r'$x(t)$',size=20)
plt.show()
```

```python
def G(u):
    t_span=[0,2*np.pi]
    y0=[7,u]
    sol=solve_ivp(fun,t_span,y0)
    return sol.y[0][-1]
```

```python
G=np.vectorize(G)
```

```python
U_list=np.linspace(-3,3,101)
Data=G(U_list)
plt.plot(U_list,Data,'k.')
plt.xlabel(r'$u$',size=20)
plt.ylabel(r'$G(u)$',size=20)
plt.show()
```

```python
U_list=np.linspace(-1.5,-1.2,101)
Data=G(U_list)
plt.plot(U_list,Data,'k.')
plt.xlabel(r'$u$',size=20)
plt.ylabel(r'$G(u)$',size=20)
plt.show()
```

```python
U_list=np.linspace(-1.35,-1.33,101)
Data=G(U_list)
plt.plot(U_list,Data,'k.')
plt.xlabel(r'$u$',size=20)
plt.ylabel(r'$G(u)$',size=20)
plt.show()
```

```python
a=-2
b=-1
for n in range(12):
    c=(a+b)/2
    if G(a)*G(c)<0:
        b=c
    else:
        a=c
    print('Iteration n:',n,'-->',c,G(c))
```
