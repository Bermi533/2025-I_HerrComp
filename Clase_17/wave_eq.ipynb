```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
delta_t=0.001
v=1.0
L=1.0
t_max=10
T=np.arange(0,t_max+delta_t,delta_t)
delta_x=v*delta_t
X=np.arange(0,L+delta_x,delta_x)
```

```python
A,B=np.meshgrid(T,X)
Psi_sol=np.zeros(A.shape)
Psi_sol.fill(0)
```

```python
#Condiciones de frontera
Psi_sol[::,0]=np.sin(X*2*np.pi/L) #Para todo x, t=0

for i in range(1,len(X)-1):
        Psi_sol[i,1]=(1.0/2)*(Psi_sol[i+1,0]+Psi_sol[i-1,0])


for j in range(1,len(T)-1):
    for i in range(1,len(X)-1): #Aquí implícitamente se incluyen las condiciones de frontera en x
        Psi_sol[i,j+1]=Psi_sol[i+1,j]+Psi_sol[i-1,j]-Psi_sol[i,j-1]

plt.imshow(Psi_sol, extent=(min(T),max(T),min(X),max(X)),aspect=5)
plt.colorbar().set_label(label=r'$\Psi(x,y)$',size=20)
plt.xlabel(r'$t$',size=20)

plt.ylabel(r'$x$',size=20)
plt.show()
```

```python
for j in range(0,len(T),10):
    plt.plot(X,Psi_sol[::,j])
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$\Psi(x,t)$',size=20)
plt.show()
```
