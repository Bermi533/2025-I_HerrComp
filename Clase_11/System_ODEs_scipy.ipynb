```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci  
```

```python
def f(t,y):
    return -2*y

t_span = np.linspace(0, 5, 100)
y0 = [1]

sol = sci.integrate.solve_ivp(f, (0,5), y0, t_eval=t_span)

plt.plot(sol.t, sol.y[0], label='Solución numérica')
plt.plot(sol.t, np.exp(-2*sol.t), '--', label='Solución exacta')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.title('Solución de dy/dt = -2y con solve_ivp')
plt.grid(True)
plt.show()
```
