```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m=1
gamma=0.1
k=1
F0=1
omega=2*np.pi

def f(t,x,y):
    return y

def g(t,x,y):
    return 1.0/m*(F0*np.cos(omega*t)-k*x-gamma*y)
```

```python
T=[]
X=[]
Y=[]

t=0
x=0
y=0

h=0.01
while t<=100:
    T.append(t)
    X.append(x)
    Y.append(y)
    k1x=f(t,x,y)
    k1y=g(t,x,y)
    k2x=f(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
    k2y=g(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
    k3x=f(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
    k3y=g(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
    k4x=f(t+h,x+k3x*h,y+k3y*h)
    k4y=g(t+h,x+k3x*h,y+k3y*h)
    x+=(1.0/6)*(k1x+2*k2x+2*k3x+k4x)*h
    y+=(1.0/6)*(k1y+2*k2y+2*k3y+k4y)*h
    t+=h
```

```python
plt.plot(T,X,'k-',label=r'$x(t)$')
plt.plot(T,Y,'r--',label=r'$v(t)$')
plt.xlabel('tiempo')
#plt.ylabel('Población')
plt.legend(loc=(1.05,0.8))
plt.show()
```

```python
def Oscillator(tmax,h):
    T=[]
    X=[]
    Y=[]
    t=0
    x=0
    y=0
    while t<=tmax:
        T.append(t)
        X.append(x)
        Y.append(y)
        k1x=f(t,x,y)
        k1y=g(t,x,y)
        k2x=f(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
        k2y=g(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
        k3x=f(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
        k3y=g(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
        k4x=f(t+h,x+k3x*h,y+k3y*h)
        k4y=g(t+h,x+k3x*h,y+k3y*h)
        x+=(1.0/6)*(k1x+2*k2x+2*k3x+k4x)*h
        y+=(1.0/6)*(k1y+2*k2y+2*k3y+k4y)*h
        t+=h
    return T,X,Y
```

```python
omega=1
T,X,Y=Oscillator(100,0.001)
plt.plot(T,X)
plt.show()
```

```python
OmegaList=np.arange(0,2,0.01)
Max_X=[]
for omega in OmegaList:
    T,X,Y=Oscillator(100,0.01)
    X=np.array(X)
    Max_X.append(max(abs(X)))
```

```python
plt.plot(OmegaList,Max_X,'k.')
plt.xlabel(r'$\omega\,[Hz]$',size=20)
plt.ylabel(r'$|x|_{max}\,[m]$',size=20)
plt.show()
```
