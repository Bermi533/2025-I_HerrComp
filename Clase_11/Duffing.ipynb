```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta=0.1*np.pi
alpha=0
beta=4*np.pi**2
gamma=30*np.pi**2
omega=2*np.pi

def f(t,x,y):
    return y

def g(t,x,y):
    return gamma*np.cos(omega*t)-alpha*x-delta*y-beta*x**3
```

```python
T=[]
X=[]
Y=[]

t=0
x=1
y=0

h=0.01
while t<=20:
    T.append(t)
    X.append(x)
    Y.append(y)
    k1x=f(t,x,y)
    k1y=g(t,x,y)
    k2x=f(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
    k2y=g(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
    k3x=f(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
    k3y=g(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
    k4x=f(t+h,x+k3x*h,y+k3y*h)
    k4y=g(t+h,x+k3x*h,y+k3y*h)
    x+=(1.0/6)*(k1x+2*k2x+2*k3x+k4x)*h
    y+=(1.0/6)*(k1y+2*k2y+2*k3y+k4y)*h
    t+=h
```

```python
plt.plot(T,X,'k-',label=r'$x(t)$')
#plt.plot(T,Y,'r--',label=r'$v(t)$')
plt.xlabel('tiempo')
#plt.ylabel('Población')
plt.legend(loc=(1.05,0.8))
plt.show()
```

```python
Xp=[]
Yp=[]

t=0
x=1
y=0

h=0.01
count=0
while t<=100000:
    if count%100==0:
        #print(t)
        Xp.append(x)
        Yp.append(y)
        count=0
    k1x=f(t,x,y)
    k1y=g(t,x,y)
    k2x=f(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
    k2y=g(t+0.5*h,x+0.5*k1x*h,y+0.5*k1y*h)
    k3x=f(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
    k3y=g(t+0.5*h,x+0.5*k2x*h,y+0.5*k2y*h)
    k4x=f(t+h,x+k3x*h,y+k3y*h)
    k4y=g(t+h,x+k3x*h,y+k3y*h)
    x+=(1.0/6)*(k1x+2*k2x+2*k3x+k4x)*h
    y+=(1.0/6)*(k1y+2*k2y+2*k3y+k4y)*h
    t+=h
    count+=1
```

```python
plt.plot(Xp,Yp,'k.',markersize=0.1,alpha=0.8)
plt.xlabel(r'$x$',size=20)
plt.ylabel(r'$\dot{x}$',size=20)
plt.show()
```
