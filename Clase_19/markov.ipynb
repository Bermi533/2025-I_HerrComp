```python
State=[]
for val in Data:
    if val in vocals:
        State.append(vocals.index(val))
```

```python
plt.plot(State)
plt.show()
```

```python
val, freq = np.unique(State,return_counts=True)
plt.plot(val,freq,'k.')
plt.xticks(list(range(5)),vocals)
plt.show()
```

```python
matC=np.zeros([5,5])
for i in range(len(State)-1):
    matC[State[i]][State[i+1]]+=1
```

```python
matw=np.zeros([5,5])
for i in range(5):
    matw[i]=matC[i]/sum(matC[i])
```
